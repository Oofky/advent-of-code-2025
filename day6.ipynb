```python
# Part 1
import re

with open("day6_1_input.txt") as file:
    list_of_lines = list(file)

# Set up list_of_num_lists
list_of_num_lines = list_of_lines[:-1]
list_of_num_lists = []
for line in list_of_num_lines:
    list_of_num_lists.append(re.split(r"\s+", line.strip()))

# Set up list_of_ops
list_of_ops = re.split(r"\s+", list_of_lines[-1].strip())

grand_total = 0

for problem_idx in range(len(list_of_ops)):
    op = list_of_ops[problem_idx]
    if op == "+":
        answer = 0
        for num_list in list_of_num_lists:
            answer += int(num_list[problem_idx])
    elif op == "*":
        answer = 1
        for num_list in list_of_num_lists:
            answer *= int(num_list[problem_idx])
    grand_total += answer
    
print(grand_total)
```

```python
# Part 2
import re

with open("day6_1_input.txt") as file:
    list_of_lines = list(file)

list_of_num_lines = list_of_lines[:-1]
list_of_ops = re.split(r"\s+", list_of_lines[-1].strip())

grand_total = 0
curr_op_idx = 0
curr_ans = 0 if list_of_ops[curr_op_idx] == "+" else 1

for i in range(len(list_of_num_lines[0])):
    num = ""
    for line_count in range(len(list_of_num_lines)):
        char = list_of_num_lines[line_count][i]
        if char.isdigit():
            num += char
    
    if num == "":
        try:
            grand_total += curr_ans # will run without fail
            curr_op_idx += 1
            curr_ans = 0 if list_of_ops[curr_op_idx] == "+" else 1 # errors at the last loop
        except:
            continue
    else:
        if list_of_ops[curr_op_idx] == "+":
            curr_ans += int(num)
        elif list_of_ops[curr_op_idx] == "*":
            curr_ans *= int(num)
            
print(grand_total)
```
