```python
# Part 1
with open("day5_1_input.txt") as file:
    fresh_str, ingr_str = file.read().split("\n\n")

fresh_lst, ingr_lst = fresh_str.split("\n"), ingr_str.split("\n")
fresh_start, fresh_end = [], []

for fresh_range in fresh_lst:
    try:
        start, end = fresh_range.split("-")
        fresh_start.append(int(start))
        fresh_end.append(int(end))
    except: # account for invalid values that cannot int()
        continue
    
count = 0

def is_Fresh(ingr):
    for i in range(len(fresh_start)):
        if ingr >= fresh_start[i] and ingr <= fresh_end[i]:
            return True
    return False
    
for ingr in ingr_lst:
    try:
        if is_Fresh(int(ingr)):
            count += 1
    except: # account for invalid values that cannot int(), such as ""
        continue
            
print(count)
```

```python
# Part 2
with open("day5_1_input.txt") as file:
    fresh_str = file.read().split("\n\n")[0]

fresh_lst = fresh_str.split("\n")
fresh_lst.sort(key = lambda e : int(e.split("-")[0]))

# Merge overlapping ranges
merged_ranges = []

def str_to_list(range_str):
    start, end = range_str.split("-")
    return [int(start), int(end)]

for fresh_range in fresh_lst:
    fresh_range = str_to_list(fresh_range)
    
    if len(merged_ranges) == 0:
        merged_ranges.append(fresh_range)
        continue
        
    fresh_start, fresh_end = fresh_range
    prev_start, prev_end = merged_ranges[-1] # fresh_start >= prev_start
    
    if fresh_end <= prev_end: # fresh range is contained in prev range
        continue
    elif fresh_start > prev_end: # no overlap at all
        merged_ranges.append(fresh_range)
    else: # a bit of overlap
        merged_ranges[-1] = [prev_start, fresh_end]

        
# Count fresh ingredients
fresh_count = 0

for merged_range in merged_ranges:
    start, end = merged_range
    fresh_count += end - start + 1
    
print(fresh_count)
```
