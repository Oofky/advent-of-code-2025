```python
# Part 1
with open("day5_1_input.txt") as file:
    fresh_str, ingr_str = file.read().split("\n\n")

fresh_lst, ingr_lst = fresh_str.split("\n"), ingr_str.split("\n")
fresh_start, fresh_end = [], []

for fresh_range in fresh_lst:
    try:
        start, end = fresh_range.split("-")
        fresh_start.append(int(start))
        fresh_end.append(int(end))
    except: # account for invalid values that cannot int()
        continue
    
count = 0

def is_Fresh(ingr):
    for i in range(len(fresh_start)):
        if ingr >= fresh_start[i] and ingr <= fresh_end[i]:
            return True
    return False
    
for ingr in ingr_lst:
    try:
        if is_Fresh(int(ingr)):
            count += 1
    except: # account for invalid values that cannot int(), such as ""
        continue
            
print(count)
```

```python
# Part 2
with open("day5_1_input.txt") as file:
    fresh_str = file.read().split("\n\n")[0]

fresh_lst = fresh_str.split("\n")
fresh_start, fresh_end = [], [] # all ranges do not overlap. To keep track of processed ranges

count = 0

"""
Cases:
1. new range is completely unique -> append the new range and add to count (end - start + 1)
2. new range is completely repeated / new range = old range -> do nothing
3. new range's starting is repeated -> change the old range's end and add to count (new end - old end)
4. new range's ending is repeated -> change the old range's start and add to count (old start - new start)
5. new range contains old range -> change the old range to new range and add to count(new end - old end + old start - new start)
"""

for fresh_range in fresh_lst:
    new_start, new_end = fresh_range.split("-")
    new_start, new_end = int(new_start), int(new_end)
    
    completely_unique = True
    
    for r in range(len(fresh_start)):
        old_start, old_end = fresh_start[r], fresh_end[r]
        print(fresh_range, new_start, new_end, old_start, old_end)
        if new_start >= old_start: # completely unique / completely repeated / start repeated
            if new_end <= old_end: # completely repeated
                print("completely repeated")
                completely_unique = False
                break
            elif new_start > old_end: # completely unique
                continue
            else: # start repeated
                print("start repeated")
                fresh_end[r] = new_end
                count += new_end - old_end
                completely_unique = False
        else: # completely unique / end repeated / new contains old
            if new_end < old_start: # completely unique
                continue
            elif new_end > old_end: # new contains old
                print("new contains old")
                fresh_start[r] = new_start
                fresh_end[r] = new_end
                count += new_end - old_end + old_start - new_start
                completely_unique = False
                break 
            else: # end repeated
                print("end repeated")
                fresh_start[r] = new_start
                count += old_start - new_start
                completely_unique = False
        
    if completely_unique:
        print("completely unique")
        fresh_start.append(new_start)
        fresh_end.append(new_end)
        count += new_end - new_start + 1
    print("AAAAAAAAAAAAAAAA", count)
            
print(count)
"""
    try:
        
    except: # account for invalid values that cannot int()
        continue
"""
```
