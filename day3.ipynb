```python
# Part 1

with open('day3_1_input.txt', 'r') as file:
    raw_list = list(file)

voltage = 0

for bank in raw_list:
    bank = bank.strip()
    bat1 = -1
    i1 = 0
    for i in range(0, len(bank) - 1): # excludes last digit
        if int(bank[i]) > bat1:
            bat1 = int(bank[i])
            i1 = i
    
    bat2 = -1
    for i in range(i1 + 1, len(bank)):
        if int(bank[i]) > bat2:
            bat2 = int(bank[i])
            
    voltage += 10 * bat1 + bat2
    
display(voltage)
```

```python
# Part 2

with open('day3_1_input.txt', 'r') as file:
    raw_list = list(file)
    
    
def findLargest(bank): # returns (max_bat: str, index_of_max: int): tuple
    max_bat = -1
    index_of_max = -1
    for i in range(len(bank)):
        curr_bat = int(bank[i])
        if curr_bat > max_bat:
            max_bat = curr_bat
            index_of_max = i
    return (str(max_bat), index_of_max)
    
    
def craftResult(bank, length_needed = 12): # returns result: string
    if length_needed == 0:
        return ""
    elif len(bank) == length_needed:
        return bank
    
    # Find largest battery and its index
    max_bat, index_of_max = findLargest(bank)
    
    left_subbank = bank[:index_of_max] # bank left of max_bat
    rest_subbank = bank[index_of_max:] # rest of the bank
    
    if len(rest_subbank) == length_needed:
        return rest_subbank
    elif len(rest_subbank) < length_needed: # need to take from front
        extra_needed = length_needed - len(rest_subbank)
        return craftResult(left_subbank, extra_needed) + rest_subbank
    else: # need to get rid of some in the back
        return max_bat + craftResult(rest_subbank[1:], length_needed - 1)

    
voltage = 0

for bank in raw_list:
    voltage += int(craftResult(bank.strip()))
    
print(voltage)
```
