```python
# Part 1
with open("day7_1_input.txt") as file:
    diagram = file.read().split("\n")
    
S_idx = diagram[0].find("S")
diagram[1] = diagram[1][:S_idx] + "|" + diagram[1][S_idx + 1:]

split_count = 0

for line_i in range(2, len(diagram)):
    curr_line = diagram[line_i]
    prev_line = diagram[line_i - 1]
    for char_i in range(len(curr_line)):
        char = curr_line[char_i]
        if prev_line[char_i] == "|":
            if char == "^":
                split_count += 1
                diagram[line_i + 1] = diagram[line_i + 1][:char_i - 1] + "|" + \
                                        diagram[line_i + 1][char_i] + \
                                        "|" + diagram[line_i + 1][char_i + 2:]
            else:
                diagram[line_i + 1] = diagram[line_i + 1][:char_i] + "|" + diagram[line_i + 1][char_i + 1:]
            
print(split_count)

with open("day7_2_diagram.txt", "w") as file:
    for line in diagram:
        file.write(line + "\n")
```

```python
# Part 2
with open("day7_2_diagram.txt") as file:
#     diagram = file.read().strip().split("\n")
    a = """.......S.......
.......|.......
......|^|......
......|.|......
.....|^|^|.....
.....|.|.|.....
....|^|^|^|....
....|.|.|.|....
...|^|^|||^|...
...|.|.|||.|...
..|^|^|||^|^|..
..|.|.|||.|.|..
.|^|||^||.||^|.
.|.|||.||.||.|.
|^|^|^|^|^|||^|
|.|.|.|.|.|||.|"""
    diagram = a.strip().split("\n")
    S_idx = diagram[0].find("S")
    
timelines = 1

def traverse(line_num, i):
    global timelines
    if line_num >= len(diagram):
        return
    line = diagram[line_num]
    if line[i] == "|":
        traverse(line_num + 2, i)
    else:
        extra_timelines = 0
        if i - 1 >= 0 and line[i - 1] == "|":
            extra_timelines += 1
            traverse(line_num + 2, i - 1)
        if i + 1 < len(diagram) and line[i + 1] == "|":
            extra_timelines += 1
            traverse(line_num + 2, i + 1)
            
        timelines += 0 if extra_timelines == 0 else extra_timelines -1
        
traverse(1, S_idx)
print(timelines)
```

```python
# Part 2
with open("day7_2_diagram.txt") as file:
    diagram = file.read().strip().split("\n")
    S_idx = diagram[0].find("S")

# Set up memoization
mem = []
depth = len(diagram) - 2 # last two lines not needed
width = len(diagram[0])
for i in range(depth):
    l = []
    for j in range(width):
        l.append(None)
    mem.append(l)
    
    
# Wishful thinking
def count_ways(parent_line_n, parent_i):
    child_line_n = parent_line_n + 2
    child_line = diagram[child_line_n]
    
    if mem[parent_line_n][parent_i] is not None:
        return mem[parent_line_n][parent_i]
    
    result = 0
    
    if child_line_n == len(diagram) - 1: # child is at the end
        if child_line[parent_i] == "|":
            result = 1
        else:
            if child_line[parent_i - 1] == "|":
                result += 1
            if child_line[parent_i + 1] == "|":
                result += 1
    else:   
        if child_line[parent_i] == "|":
            result = count_ways(child_line_n, parent_i)
        else:
            result = (0 if child_line[parent_i - 1] != "|" else count_ways(child_line_n, parent_i - 1)) + \
                    (0 if child_line[parent_i + 1] != "|" else count_ways(child_line_n, parent_i + 1))
            
    mem[parent_line_n][parent_i] = result
    return result

        
count_ways(1, S_idx)
```
