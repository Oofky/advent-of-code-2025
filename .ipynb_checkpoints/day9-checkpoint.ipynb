```python
# Part 1

with open("day9_1_input.txt") as file:
    list_of_tile_str = file.read().strip().split("\n")
    
largestArea = 0
    
for a in range(len(list_of_tile_str)):
    xA, yA = list_of_tile_str[a].split(",")
    xA, yA = int(xA), int(yA)
    for b in range(a + 1, len(list_of_tile_str)):
        xB, yB = list_of_tile_str[b].split(",")
        xB, yB = int(xB), int(yB)
        area = abs(xA - xB + 1) * abs(yA - yB + 1) # +1 because we are counting discrete tiles, not range
        largestArea = max(area, largestArea)
        
print(largestArea)
```
