```python
# Part 1

with open('day2_1_input.txt', 'r') as file:
    contents = list(file)[0]
    list_of_ranges = contents.strip().split(",")
    list_of_IDs = []
    
    for i in range(len(list_of_ranges)):
        a, b = list_of_ranges[i].split("-")
        a, b = int(a), int(b)
        for num in range(a, b + 1):
            list_of_IDs.append(str(num))

total = 0

for i in range(len(list_of_IDs)):
    s = list_of_IDs[i]
    length = len(s)
    if length % 2 == 0 and s[:length//2] == s[length//2:]: # length definitely must be even
        total += int(s)
        
print(total)
```

```python
# Part 2 (I can tell this is very inefficient...)
import math

with open('day2_1_input.txt', 'r') as file:
    contents = list(file)[0]
    list_of_ranges = contents.strip().split(",")
    list_of_IDs = []
    
    for i in range(len(list_of_ranges)):
        a, b = list_of_ranges[i].split("-")
        a, b = int(a), int(b)
        for num in range(a, b + 1):
            list_of_IDs.append(str(num))

total = 0

for i in range(len(list_of_IDs)):
    s = list_of_IDs[i]
    length = len(s)
    for n in range(1, math.ceil(length / 2) + 1):
        if length % n == 0 and length // n >= 2: # eg if num is of length 9 and n = 3 AND repeats at least twice
            if s == s[:n] * (length // n):
                total += int(s)
                break # to avoid double counting the same string (eg 9999 can be 9 9 9 9 and 99 99)
        
print(total)
```
