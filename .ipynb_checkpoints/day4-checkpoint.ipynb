```python
# Part 1

with open('day4_1_input.txt', 'r') as file:
    raw_list = list(file)
    
def isInRange(l, y, x):
    return y >= 0 and x >= 0 and y < len(l) and x < len(l[0])
    
def isAccessible(l, row, col): # checks the 9 positions, including the @ at l[row][col]
    num = -1 # offset by 1 to account for the middle position
    for y in range(row - 1, row + 2):
        for x in range(col - 1, col + 2):
            if isInRange(l, y, x) and l[y][x] == '@':
                num += 1
            if num >= 4:
                return False
    return True

count = 0
    
for r in range(len(raw_list)):
    row = raw_list[r].strip()
    for c in range(len(row)):
        col = row[c]
        if col == '@' and isAccessible(raw_list, r, c):
            count += 1
            
print(count)                     
```

```python
# Part 2

hasRemoved = True # start with True to start the loop
count = 0

while hasRemoved:
    hasRemoved = False
    
    for r in range(len(raw_list)):
        row = raw_list[r].strip()
        for c in range(len(row)):
            col = row[c]
            if col == '@' and isAccessible(raw_list, r, c):
                count += 1
                hasRemoved = True # this is not the last loop, will need to loop again
                raw_list[r] = raw_list[r][:c] + 'x' + raw_list[r][c + 1:]
                
print(count)
print(raw_list)
```
