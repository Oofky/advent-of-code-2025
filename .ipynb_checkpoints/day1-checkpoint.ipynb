```python
# Part 1
with open('day1_1_input.txt', 'r') as file:
    rot_list = list(file)
    
    for i in range(len(rot_list)):
        entire = rot_list[i].strip()
        direction = entire[0]
        num = int(entire[1:])
        rot_list[i] = [direction, num]
        
    count_of_zero = 0
    dial = 50
    
    for i in range(len(rot_list)):
        p = rot_list[i]
        
        if p[0] == "L":
            dial -= p[1]
        elif p[0] == "R":
            dial += p[1]
            
        ## correct dial
        dial = dial % 100
            
        ## count zeroes
        if dial == 0:
            count_of_zero += 1
            
        print(f'{dial}, {p}')
            
    print(count_of_zero)
```

```python
# Part 2 (snap to zero)

with open('day1_1_input.txt', 'r') as file:
    rot_list = list(file)
    
    for i in range(len(rot_list)):
        entire = rot_list[i].strip()
        direction = entire[0]
        num = int(entire[1:])
        rot_list[i] = [direction, num]
        
    count_of_zero = 0
    dial = 50
    
    for i in range(len(rot_list)):
        p = rot_list[i]
        
        # can i snap to zero?
        if p[0] == "L":
            if p[1] >= dial: # yes i can snap to zero
                used = dial
                count_of_zero += 1
                dial = 0
                # how many times can i still snap to zero?
                if p[1] - used > 100:
                    count_of_zero += (p[1] - used) // 100 + 1
                dial -= (p[1] % 100)
                dial = dial % 100
            else: # no i cannot snap to zero
                dial -= p[1]
        else:
            if dial + p[1] > 99: # yes i can snap to zero
                used = 99 - dial
                count_of_zero += 1
                dial = 0
                # how many times can i still snap to zero?
                if p[1] - used > 100:
                    count_of_zero += (p[1] - used) // 100
                dial += (p[1] % 100)
                
    print(count_of_zero)
```

```python
with open('day1_1_input.txt', 'r') as file:
    rot_list = list(file)
    
    for i in range(len(rot_list)):
        entire = rot_list[i].strip()
        direction = entire[0]
        num = int(entire[1:])
        rot_list[i] = [direction, num]
        
    count_of_zero = 0
    dial = 50
    
    for i in range(len(rot_list)):
        p = rot_list[i]
        if p[0] == 'L':
            dial -= p[1]
        else:
            dial += p[1]
        if dial >= 100 or dial <= -100:
            count_of_zero += abs(dial) // 100
            if abs(dial) % 100 < 0:
                count_of_zero += 1
        else:
            if dial <= 0:
                count_of_zero += 1
        dial = dial % 100

    print(count_of_zero)
```
