```python
# Part 1

# Init list of junction boxes, list of connections, list of circuits
list_of_jbox = []
list_of_conn = []
list_of_circ = []


# Populate list of junction boxes
with open("day8_1_input.txt") as file:
    list_of_jbox_str = file.read().strip().split("\n")
    
def jbox_str_into_tuple(jbox_str):
    x, y, z = jbox_str.split(",")
    return (int(x), int(y), int(z))

for jbox_str in list_of_jbox_str:
    list_of_jbox.append(jbox_str_into_tuple(jbox_str))
    
    
# Populate list of connections
def calculate_dist(jbox_A, jbox_B):
    xA, yA, zA = jbox_A
    xB, yB, zB = jbox_B
    return ((xA-xB)**2 + (yA-yB)**2 + (zA-zB)**2) ** 0.5
    
class Conn():
    def __init__(self, jbox_A, jbox_B):
        self.jbox_A = jbox_A
        self.jbox_B = jbox_B
        self.dist = calculate_dist(jbox_A, jbox_B)
    
for i in range(len(list_of_jbox_str)):
    for j in range(i + 1, len(list_of_jbox_str)):
        list_of_conn.append(Conn(list_of_jbox[i], list_of_jbox[j]))
        
list_of_conn.sort(key = lambda e : e.dist)


# Populate list of circuits
for jbox in list_of_jbox:
    list_of_circ.append([jbox])
    
added_count = 0

for conn in list_of_conn:
    if added_count >= 1000:
        break
    
    jbox_A = conn.jbox_A
    jbox_B = conn.jbox_B
    
    # Find the index of the circuit that A and B are in
    A_is_in = None
    B_is_in = None
    
    for i in range(len(list_of_circ)):
        if A_is_in is not None and B_is_in is not None: # already found
            break
        
        circ = list_of_circ[i]
        if jbox_A in circ:
            A_is_in = i
        if jbox_B in circ:
            B_is_in = i
            
    if A_is_in != B_is_in: # not already in same circuit
        circ_A = list_of_circ[A_is_in]
        circ_B = list_of_circ[B_is_in]
        merged = circ_A + circ_B
        
        list_of_circ.remove(circ_A)
        list_of_circ.remove(circ_B)
        list_of_circ.append(merged)
        
    added_count += 1


list_of_circ.sort(key = lambda e : len(e), reverse = True)

print(len(list_of_circ[0]) * len(list_of_circ[1]) * len(list_of_circ[2]))
```
