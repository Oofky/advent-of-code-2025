```python
# Part 1

with open("day11_1_input.txt") as file:
    line_list = file.read().strip().split("\n")

cable_dict = {}
for line in line_list:
    device, outputs = line.split(":")
    outputs = outputs.strip().split(" ")
    cable_dict[device] = outputs
    
paths = 0

def traverse(device):
    global paths
    if device == "out":
        paths += 1
    else:
        outputs = cable_dict[device]
        for output in outputs:
            traverse(output)
            
traverse("you")
print(paths)
```
